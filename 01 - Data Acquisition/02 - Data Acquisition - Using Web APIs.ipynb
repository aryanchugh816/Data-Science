# 2. Facebook API

### Requests Library
* Most Popular Python Library
* Wrapper over Urllib
* Works well with Python 3

### Using Requests Library

```python
import requests
```

```python
# url = "http://graph.facebook.com/<user_id(numeric)>/picture?type=large"
# everything after question mark('?') if a query or number of attributes and authentication code if needed
url = "http://graph.facebook.com/4/picture?type=large"
```

```python
r = requests.get(url)
```

```python
print(r)
print(type(r.content))
# we get a response object hence url is correct and we can get data from it
```

```python
# To get the content of the url
print(r.content)

# all this content is in binary and we will have to convert it to get a real picture
```

```python
with open("./Data/sample_pic.jpeg", 'wb') as f:  # 'wb' stands for write-binary
    f.write(r.content)
    
# This will create a jpeg file from the binary content we had
```

```python
# We will visualize image using opencv and matplotlib
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./Data/sample_pic.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

# 4. Image Scrapping
* We will use BeautifulSoup(of bs4 library) instead of any specific api
* Basically we will be scrapping images from html code of any website
* While BeautifulSoup or any other parser or librarywe will receive the image in binary format
 and will need to store it to a .jpeg/.jpg file using 'wb'(write binary) mode

```python
import bs4
import requests

# Do the usual stuff of making a url and getting a response from it
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=1"
```

```python
response = requests.get(url)
```

```python
response.content
```

```python
# Now we need to create a BeautifulSoup object which will help us parse the html content easier
soup = bs4.BeautifulSoup(response.content, 'html.parser') # We can specify which type of parser we want but default parser is of html

# Now we need to inspect element the url page and see which tag contains the url of images
```

```python
div_element = soup.findAll('div')
print(len(div_element))
```

```python
soup
```

```python
divs = soup.findAll('div', {'class' : 'row', 'id' : 'all_quotes'})
```

```python
anchor = divs[0].findAll('a')
```

```python
len(divs)
# Because there is only one div with that specific class and id name
```

```python
anchor[0]
```

```python
anchor[1]
```

```python
anchor[2]
# As we can see even numbered indexed anchors contain image tag hence will sort only them
```

```python
len(anchor)
```

```python
# here we have anchor tags with image tags
for i in range(0,64,2):
    print(anchor[i])
```

```python
anchor[0].img.attrs['src']
```

```python
anchor[0].img.attrs
# .attrs is used in BeautifulSoup to return a dictionary of attributes of that tag
```

```python
# Now we write all the binary data into a .jpg file after retreiving it from website
for i in range(0,64,2):
    with open('Data/Scraped Images/Inspirational quote no.{}.jpg'.format(i), 'wb') as f:
        img_url = anchor[i].img.attrs['src']
        response = requests.get(img_url)
        f.write(response.content)
        
# This will form different .jpg files and images will be stored in the current file directory
```
